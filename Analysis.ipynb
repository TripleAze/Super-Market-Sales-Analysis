```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Load dataset
df = pd.read_csv("Sales Super Market DF")
```

```python
#Data Overview
df.head()
```

```python
# Dataset Info
df.info()
```

```python
# Missing Values
df.isnull().sum()
```

```python
# Duplicate Rows
df.duplicated().sum()
```

```python
# Descriptive Statistics
df.describe()
```

```python
# Categorical column unique values
categorical_columns = ['Branch', 'City', 'Customer type', 'Gender', 'Product line', 'Payment']
for col in categorical_columns:
    print(f"{col}: {df[col].unique()}")
```

```python
# Quick overview of key columns
key_columns = ['Sales', 'Rating', 'gross income']
print("\nQuick Overview of Key Columns:")
print(df[key_columns].describe())
```

```python
# Converting date column to the right column 
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
```

```python
df['Time'] = df['Time'].str[:5]  # Truncate to keep only hours and minutes
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M').dt.hour
```

```python
# Sales and gross income by branch
branch_sales = df.groupby('Branch').agg(
    total_sales=('Sales', 'sum'),
    total_gross_income=('gross income', 'sum')
).reset_index().sort_values(by='total_sales', ascending=False)
print(branch_sales)
```

```python
# Plot: Total Sales and Gross Income by Branch
plt.figure(figsize=(10, 6))
sns.barplot(x='total_sales', y='Branch', data=branch_sales, palette='viridis')
plt.title('Total Sales by Branch')
plt.xlabel('Total Sales')
plt.ylabel('Branch')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='total_gross_income', y='Branch', data=branch_sales, palette='viridis')
plt.title('Total Gross Income by Branch')
plt.xlabel('Gross Income')
plt.ylabel('Branch')
plt.show()
```

```python
# Total sales by product line
product_line_sales = df.groupby('Product line').agg(
    total_sales=('Sales', 'sum')
).reset_index().sort_values(by='total_sales', ascending=False)
print(product_line_sales)
```

```python
# Plot: Total Sales by Product Line
plt.figure(figsize=(10, 6))
sns.barplot(x='total_sales', y='Product line', data=product_line_sales, palette='Set2')
plt.title('Total Sales by Product Line')
plt.xlabel('Total Sales')
plt.ylabel('Product Line')
plt.show()
```

```python
# Average total sales by customer type
customer_type_sales = df.groupby('Customer type').agg(
    avg_total_sales=('Sales', 'mean')
).reset_index()
print(customer_type_sales)
```

```python
# Plot: Average Sales by Customer Type
plt.figure(figsize=(8, 5))
sns.barplot(x='Customer type', y='avg_total_sales', data=customer_type_sales, palette='Blues')
plt.title('Average Sales by Customer Type')
plt.xlabel('Customer Type')
plt.ylabel('Average Total Sales')
plt.show()
```

```python
# Average total sales and ratings by gender
gender_sales_ratings = df.groupby('Gender').agg(
    avg_total_sales=('Sales', 'mean'),
    avg_rating=('Rating', 'mean')
).reset_index()
print(gender_sales_ratings)
```

```python
# Plot: Average Sales and Ratings by Gender
plt.figure(figsize=(8, 5))
sns.barplot(x='Gender', y='avg_total_sales', data=gender_sales_ratings, palette='Pastel1')
plt.title('Average Sales by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Total Sales')
plt.show()

plt.figure(figsize=(8, 5))
sns.barplot(x='Gender', y='avg_rating', data=gender_sales_ratings, palette='Pastel1')
plt.title('Average Ratings by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Rating')
plt.show()
```

```python
# Payment method analysis
payment_method_count = df['Payment'].value_counts().reset_index()
payment_method_count.columns = ['Payment', 'Frequency']
print(payment_method_count)
```

```python
# Plot: Payment Method Frequency
plt.figure(figsize=(8, 5))
sns.barplot(x='Payment', y='Frequency', data=payment_method_count, palette='coolwarm')
plt.title('Payment Method Frequency')
plt.xlabel('Payment Method')
plt.ylabel('Frequency')
plt.show()
```

```python
# Average sales by payment method
payment_sales_avg = df.groupby('Payment').agg(
    avg_total_sales=('Sales', 'mean')
).reset_index()
print(payment_sales_avg)
```

```python
# Plot: Average Sales by Payment Method
plt.figure(figsize=(8, 5))
sns.barplot(x='Payment', y='avg_total_sales', data=payment_sales_avg, palette='coolwarm')
plt.title('Average Sales by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Average Total Sales')
plt.show()
```

```python
# Daily total sales
daily_sales = df.groupby('Date').agg(
    total_sales=('Sales', 'sum')
).reset_index().sort_values(by='total_sales', ascending=False)
print(daily_sales)
```

```python
# Plot: Total Sales by Date
plt.figure(figsize=(15, 8))
sns.lineplot(x='Date', y='total_sales', data=daily_sales, marker='o')
plt.title('Total Sales by Date')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Hourly transactions
hourly_transactions = df.groupby('Hour').agg(
    transaction_count=('Invoice ID', 'count')
).reset_index()
print(hourly_transactions)
```

```python
# Plot: Transactions by Hour
plt.figure(figsize=(8, 5))
sns.barplot(x='Hour', y='transaction_count', data=hourly_transactions, palette='coolwarm')
plt.title('Number of Transactions by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Transaction Count')
plt.show()
```

```python
# Customer type transaction counts
transaction_count_customer_type = df.groupby('Customer type').agg(
    transaction_count=('Invoice ID', 'count')
).reset_index()
print(transaction_count_customer_type)
```

```python
# Plot: Transaction Counts by Customer Type
plt.figure(figsize=(8, 5))
sns.barplot(x='Customer type', y='transaction_count', data=transaction_count_customer_type, palette='coolwarm')
plt.title('Transaction Counts by Customer Type')
plt.xlabel('Customer Type')
plt.ylabel('Transaction Count')
plt.show()
```

```python
# Rating vs. Sales scatterplot
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Rating', y='Sales', data=df, alpha=0.7, color='purple')
plt.title('Relationship Between Rating and Total Sales')
plt.xlabel('Customer Rating')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Average rating by product line
product_line_rating = df.groupby('Product line').agg(
    avg_rating=('Rating', 'mean')
).reset_index()
print(product_line_rating)
```

```python
# Plot: Average Ratings by Product Line
plt.figure(figsize=(10, 6))
sns.barplot(x='avg_rating', y='Product line', data=product_line_rating, palette='viridis')
plt.title('Average Customer Ratings by Product Line')
plt.xlabel('Average Rating')
plt.ylabel('Product Line')
plt.show()
```

```python
# Payment preferences by customer type
payment_customer_type = df.groupby(['Payment', 'Customer type']).size().reset_index(name='Count')
print(payment_customer_type)
```

```python
# Plot: Payment Preferences by Customer Type
plt.figure(figsize=(10, 6))
sns.barplot(x='Payment', y='Count', hue='Customer type', data=payment_customer_type, palette='coolwarm')
plt.title('Payment Method Preferences by Customer Type')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.legend(title='Customer Type')
plt.show()
```

```python
# Payment preferences by product line
payment_product_line = df.groupby(['Payment', 'Product line']).size().reset_index(name='Count')
print(payment_product_line)
```

```python
# Plot: Payment Preferences by Product Line
plt.figure(figsize=(12, 8))
sns.barplot(x='Payment', y='Count', hue='Product line', data=payment_product_line, palette='muted')
plt.title('Payment Method Preferences by Product Line')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.legend(title='Product Line', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
